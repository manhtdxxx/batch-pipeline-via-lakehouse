```python
import pandas as pd
import warnings
from typing import Callable
import time

warnings.filterwarnings("ignore")
```

```python
def get_companies(symbols: list[str], source: str) -> list[str]:
    companies = []
    for s in symbols.to_list():
        companies.append(Company(symbol=s, source=source))
    return companies

companies_from_vci = get_companies(vn30_symbols, "VCI")
companies_from_tcbs = get_companies(vn30_symbols, "TCBS")
```

```python
def get_company_data(func: Callable[..., pd.DataFrame], companies: list[str], **kwargs) -> pd.DataFrame:
    list_df = []
    for c in companies:
        df_c = func(c, **kwargs)
        df_c["symbol"] = c.symbol
        list_df.append(df_c)

    df = pd.concat(list_df, axis=0, ignore_index=True)
    return df
```

```python
time.sleep(60)
get_company_data(Company.overview, companies_from_vci).head(5)
```

```python
time.sleep(60)
get_company_data(Company.overview, companies_from_tcbs).head(5)
```

```python
get_company_data(Company.shareholders, companies_from_vci).head(5)
```

```python
get_company_data(Company.shareholders, companies_from_tcbs).head(5)
```

```python
get_company_data(Company.officers, companies_from_vci).head(5)
```

```python
time.sleep(60)
get_company_data(Company.officers, companies_from_tcbs).head(5)
```

```python
time.sleep(60)
get_company_data(Company.news, companies_from_vci).head(3)
```
